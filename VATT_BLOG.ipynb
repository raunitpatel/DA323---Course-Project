<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 300;margin-bottom: 20px;">
  Because of its innovative approach to multimodal learning, I decided to investigate the Video-Audio-Text Transformer (VATT). I found it fascinating that a single Transformer architecture could process raw text, audio, and video data without the need for convolutional neural networks (CNNs). With the growing expectation that AI models can handle a variety of data types, VATT's self-supervised learning approach provides a scalable solution that stays clear of the drawbacks of labeled data dependency, including bias and expensive annotation.
</h5>

<h4 style="font-family: Poppins, sans-serif; font-size: 24px; font-weight: 500; margin-bottom: 20px; background-color: #6c5ce7; color: #ffffff; padding: 15px; border-radius: 10px;">
  Historical Perspective
</h4>

<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 300; margin-bottom: 20px;">
  <p  >The <strong>Video-Audio-Text Transformer (VATT)</strong>, introduced in 2021, marked a major breakthrough in multimodal learning by using a <em>convolution-free</em> Transformer architecture for self-supervised learning directly from raw video, audio, and text data. To understand its significance, it's helpful to look at the evolution of the underlying technologies.</p>

<p>The introduction of Transformers in 2017, through the seminal paper <em>Attention is All You Need</em>, revolutionized natural language processing (NLP). By replacing recurrent models with self-attention mechanisms, Transformers enabled better handling of long-range dependencies. Models like <strong>BERT</strong> and <strong>GPT</strong> demonstrated the power of large-scale pre-training, achieving state-of-the-art results across a wide range of NLP tasks.</p>

<p>Inspired by this success, Transformers began making their way into vision tasks. Early efforts used hybrid architectures that combined convolutional neural networks (CNNs) with attention mechanisms. However, the <strong>Vision Transformer (ViT)</strong> showed that pure Transformer models could match or even outperform CNNs in image recognitionprovided they were trained on large enough datasetschallenging the conventional need for convolutional biases.</p>

<p>Multimodal learning, which combines information from video, audio, and text, has been explored since the 1980s in areas like audio-visual speech recognition. The rise of Transformers brought a unified framework to this field. Models such as <strong>VideoBERT</strong>, <strong>ViLBERT</strong>, <strong>LXMERT</strong>, and <strong>CLIP</strong> advanced cross-modal pre-training. Building on this progress, <strong>VATT</strong> takes things further by processing raw multimodal inputs without relying on CNNs. It uses self-supervised contrastive learning and has achieved impressive performancefor example, <strong>82.1% top-1 accuracy on the Kinetics-400 benchmark</strong>.</p>

<p ><strong>VATT</strong> represents a shift from modality-specific architectures to a <em>general-purpose</em>, <em>convolution-free</em> Transformer approach. It even includes a <strong>modality-agnostic variant</strong> that shares weights across different data types. By leveraging large amounts of unlabeled multimodal data, VATT overcomes the biases and limitations of supervised learning, and sets new standards in video, audio, and text understandingwithout needing manually curated labels.</p>

</h5>


<h4 style="font-family: Poppins, sans-serif; font-size: 24px; font-weight: 500; margin-bottom: 20px; background-color: #6c5ce7; color: #ffffff; padding: 15px; border-radius: 10px;">
 Key Learnings
</h4>


<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 300;margin-bottom: 20px;">
  <p>Studying the Video-Audio-Text Transformer (VATT) framework, as presented in the 2021 NeurIPS paper, has provided deep insights into the potential of Transformer architectures for multimodal learning from unlabeled data. Below, I elaborate on the key takeaways from this work, focusing on its innovative approaches and implications for future research in computer vision, audio processing, and multimodal integration. </p>
<ul>
    <li><p><strong>Power of Convolution-Free Transformers Across Modalities</strong>: One of the most striking learnings from VATT is the effectiveness of convolution-free Transformer architectures in processing raw multimodal data-video (RGB frames), audio (waveforms), and text (transcripts)-without relying on Convolutional Neural Networks (CNNs). Historically, CNNs have dominated vision and audio tasks due to their inductive biases like translation invariance and locality. VATT challenges this by demonstrating that Transformers, with self-attention mechanisms, can outperform CNN-based models when trained on large-scale unlabeled data, achieving state-of-the-art results such as 82.1% top-1 accuracy on Kinetics-400 for video action recognition and 39.4% mAP on AudioSet for audio event classification. This suggests that Transformers can learn necessary priors directly from data, reducing the need for predefined architectural biases.</p></li>
    <li><p><strong>Self-Supervised Learning with Multimodal Contrastive Losses</strong>: VATT's use of self-supervised learning via multimodal contrastive losses (Noise Contrastive Estimation for video-audio pairs and Multiple Instance Learning NCE for video-text pairs) is a critical insight. By training on unlabeled datasets like HowTo100M and AudioSet, VATT aligns representations across modalities in a common semantic space, capturing cross-modal relationships without human-annotated labels. This approach addresses the limitations of supervised pre-training, such as high costs and potential biases in labeled data, and highlights the organic supervision inherent in multimodal videos-where temporal and contextual alignments between video, audio, and text serve as natural training signals. The success of this method, evidenced by competitive zero-shot text-to-video retrieval on YouCook2 and MSR-VTT, underscores the scalability of self-supervised strategies for multimodal tasks.</p></li>
        <li><p><strong>Modality-Agnostic Transformer Potential</strong>: A fascinating aspect of VATT is the exploration of a single-backbone, modality-agnostic Transformer where weights are shared across video, audio, and text modalities (with separate tokenization and projection layers). The preliminary results show that this unified model performs on par with modality-specific Transformers of slightly smaller sizes, as seen in comparable accuracies for video action recognition and audio event classification. This finding suggests a future direction for AI where a single general-purpose architecture could handle diverse data types, akin to unified language models in NLP that support multiple languages. The visualization of feature representations further supports this, showing that the modality-agnostic backbone perceives different modalities as symbols of the same underlying concept.</p></li>

<li>
    <p>
        <strong>DropToken for Computational Efficiency</strong>: VATT introduces DropToken, a novel technique to manage the quadratic computational complexity of Transformers (O(N) with respect to input tokens). By randomly dropping a portion of video and audio tokens during training (e.g., 50% drop rate in experiments), VATT enables the use of high-resolution inputs while significantly reducing training costs, with minimal impact on downstream performance. For instance, experiments show that high-resolution inputs with DropToken achieve comparable or better accuracy on Kinetics-400 than low-resolution inputs without dropping tokens. This innovation is particularly valuable for processing raw, redundant multimodal data like video and audio, making large-scale Transformer training more feasible on limited hardware.
        </p>
</li>

<li>
    <p>
    <strong>Transferability Across Domains</strong>: Another key learning is VATT's ability to transfer learned representations across domains, notably from video to image classification. Despite being pre-trained on multimodal video data, fine-tuning VATT's vision Transformer on ImageNet yields a top-1 accuracy of 78.7%, close to ViT's 79.9% (which uses supervised pre-training on large-scale image data) and far surpassing the 64.7% achieved by training the same Transformer from scratch. This demonstrates the generalizability of self-supervised multimodal pre-training, even when there is a domain gap between training (videos) and testing (images), highlighting the robustness of the learned visual priors.
        
</p>
</li>
</ul>
<p>
    In summary, VATT reshapes the understanding of multimodal learning by proving that Transformers can serve as versatile, general-purpose architectures for raw data across modalities, leveraging self-supervised techniques to bypass the constraints of supervised training. Its innovations like DropToken and the modality-agnostic approach offer practical and conceptual advancements, while its limitations highlight actionable future research directions. If you're interested in exploring specific aspects like the DropToken mechanism or potential improvements for text data quality, feel free to ask!
</p>

<h4 style="font-family: Poppins, sans-serif; font-size: 24px; font-weight: 500; margin-bottom: 20px; background-color: #6c5ce7; color: #ffffff; padding: 15px; border-radius: 10px;">
    Approach
</h4>

<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 300;margin-bottom: 20px;">
<p>There are two major settings: </p>
<ol>
    <li>
         The backbone Transformers are separate and have specific weights for each modality, and
    </li>
    <li>The Transformers share weights, namely, there is a single backbone Transformer applied to any of
 the modalities.</li>
</ol>
    <p> In either setting, the backbone extracts modality-specific representations, which are
 then mapped to common spaces to be compared with each other by contrastive losses. We describe
 each module in the following.</p>

<ol>
<li>
    <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong>Tokenization and Positional Encoding</strong></h5>
    <p>

VATT processes raw signals from video, audio, and text into vector sequences for Transformers using modality-specific tokenization and positional encoding.
<ul>
<li> <strong>Video</strong>: Raw RGB video clips ($T \times H \times W$) are split into patches of $t \times h \times w \times 3$ voxels. Each patch is linearly projected to a $d$-dimensional vector with weight $W_{vp} \in \mathbb{R}^{(t \cdot h \cdot w \cdot 3) \times d}$. Positional embeddings (temporal, horizontal, vertical) preserve spatial-temporal order using $\lceil T/t \rceil + \lceil H/h \rceil + \lceil W/w \rceil$ learnable embeddings. The positional encoding for a patch at position $(i,j,k)$ is defined as:  
$$
\mathbf{e}_{i,j,k} = \mathbf{e}_{\text{Temporal},i} + \mathbf{e}_{\text{Horizontal},j} + \mathbf{e}_{\text{Vertical},k}, 
    $$
    $$
\mathbf{E}_{\text{Temporal}} \in \mathbb{R}^{\lceil T/t \rceil \times d}, \quad
\mathbf{E}_{\text{Horizontal}} \in \mathbb{R}^{\lceil H/h \rceil \times d}, \quad
\mathbf{E}_{\text{Vertical}} \in \mathbb{R}^{\lceil W/w \rceil \times d}
$$


    
 </li>
<li><strong>Audio</strong>: Raw waveforms of length $T'$ are divided into $\lceil T'/t' \rceil$ segments of $t'$ amplitudes, projected to $d$-dimensional vectors with weight $W_{ap} \in \mathbb{R}^{t' \times d}$. Temporal order is encoded with $\lceil T'/t' \rceil$ learnable embeddings.  </li>
<li><strong>Text</strong>: Word sequences are mapped to $v$-dimensional one-hot vectors based on a vocabulary of size $v$, then projected to $d$-dimensional embeddings with weight $W_{tp} \in \mathbb{R}^{v \times d}$. Positional encodings maintain sequence order.</li>

</ul>
    </p>
    <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;">DropToken</h5>
    <p>
        VATT introduces <strong>DropToken</strong>, a simple yet effective technique to lower the computational complexity of training Transformer models. After tokenizing video and audio inputs into sequences, DropToken randomly samples a portion of these tokens, feeding only the sampled subset-rather than the full set-to the Transformer. This is critical because a Transformer's computational complexity is quadratic, $O(N^2)$, where $N$ is the number of input tokens. Reducing the input length via DropToken quadratically decreases the number of FLOPs, directly impacting training time and enabling large models to run on limited hardware.
    </p>

</li>


<p>Figure 1: <strong>Overview of the VATT architecture and the self-supervised, multimodal learning
 strategy</strong>. VATT linearly projects each modality into a feature vector and feeds it into a Transformer
 encoder. We define a semantically hierarchical common space to account for the granularity of
 different modalities and employ the Noise Contrastive Estimation (NCE) to train the model.</p>

<li>
    <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong> TheTransformer Architecture</strong></h5>
    

<p>
VATT adopts the well-established Transformer architecture, widely used in NLP, to ensure compatibility and ease of weight transfer to standard implementations, similar to Vision Transformer (ViT). This section summarizes the pipeline as described in the research paper.
</p>
<p><strong>Input Sequence Formulation</strong>:  
The input to the Transformer is a sequence of tokens formulated as:  
$$ \mathbf{z}_{\text{in}} = [\mathbf{x}_{\text{AGG}};~\mathbf{x}_0 \mathbf{W}_P;~\mathbf{x}_1 \mathbf{W}_P;~\dots;~\mathbf{x}_N \mathbf{W}_P] + \mathbf{e}_{\text{POS}}
 $$  
Here, $\mathbf{x}_N \mathbf{W}_P$ represents the sequence of input patches (or tokens) after projection, $\mathbf{x}_{\text{AGG}}$ is a learnable embedding for a special aggregation token, and $e_{POS}$ denotes positional encodings. The output corresponding to $\mathbf{x}_{\text{AGG}}$ (denoted as $\mathbf{z}_{\text{in}}$) serves as the aggregated representation of the entire sequence, used for tasks like classification and common space mapping.
</p>
<p><strong>Architecture Components</strong>:  
VATT employs standard self-attention in the Multi-Head Attention (MHA) module, GeLU activation in the MLP layer, and applies Layer Normalization before both MHA and MLP modules. For the text model specifically, positional encoding ($e_{POS}$) is omitted, and a learnable relative bias is added to each attention score in the first layer of the MHA module. This modification ensures the text model's weights are directly transferable to state-of-the-art models like T5.</p>
</li>

<li>
    <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong>Common Space Projection</strong></h5>
    <p>VATT employs common space projection and contrastive learning to train its networks using unlabeled multimodal data. This section summarizes the approach for mapping video, audio, and text representations into shared spaces for comparison, as detailed in the research paper.</p>
<p><strong>Semantically Hierarchical Common Space Mapping</strong>:  
Given a video-audio-text triplet, VATT defines a semantically hierarchical common space mapping to enable direct comparison of video-audio and video-text pairs using cosine similarity. Recognizing different levels of semantic granularity across modalities, VATT introduces multi-level projections to account for these differences as an inductive bias. The projections are formulated as:  
$$
\mathbf{z}_{v,va} = g_{v \rightarrow va}(\mathbf{z}_{\text{out}}^{\text{video}}),  \mathbf{z}_{a,va} = g_{a \rightarrow va}(\mathbf{z}_{\text{out}}^{\text{audio}}) 
$$
$$
\mathbf{z}_{t,vt} = g_{t \rightarrow vt}(\mathbf{z}_{\text{out}}^{\text{text}}), \mathbf{z}_{v,vt} = g_{v \rightarrow vt}(\mathbf{z}_{v,va})
$$
 
Here, $g_{v \rightarrow va}$ and $g_{a \rightarrow va}$ are projection heads mapping video and audio Transformer outputs to the video-audio common space $S_{va}$. Similarly, $g_{t}^{vt}$ and $g_{v}^{vt}$ project text Transformer outputs and the video embedding from $S_{va}$ to the video-text common space $S_{vt}$.</p>

<p><strong>Projection Head Design</strong>:  
The projection heads are designed with specific architectures to handle semantic granularity. For $g_{a \rightarrow va}$, $g_{t \rightarrow vt}$, and $g_{v \rightarrow vt}$, a linear projection is used. For $g_{v \rightarrow va}$, a two-layer projection with ReLU activation between layers is applied to capture more complex transformations. Batch normalization follows each linear layer to stabilize training.</p>
</li>

<li>
    <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong>Multimodal Contrastive Learning</strong></h5>


<p>VATT employs multimodal contrastive learning to train its networks end-to-end using unlabeled data, aligning representations across video, audio, and text modalities. This section summarizes the approach as detailed in the research paper. </p>

<p><strong>Noise Contrastive Estimation (NCE) for Video-Audio Pairs</strong>:  
Inspired by prior works, VATT uses Noise Contrastive Estimation (NCE) to align video-audio pairs. Positive pairs are formed by sampling corresponding video and audio streams from the same temporal location in a video, while negative pairs are created from non-matching locations. In the common space defined earlier, the NCE loss is formulated as:  
$$
\text{NCE}(\mathbf{z}_{v,va}, \mathbf{z}_{a,va}) = 
- \log\left(\frac{\exp\left(\mathbf{z}_{v,va}^\top \mathbf{z}_{a,va} / \tau \right)}{\exp\left(\mathbf{z}_{v,va}^\top \mathbf{z}_{a,va} / \tau \right) + \sum_{\mathbf{z}' \in \mathcal{N}} \exp\left({\mathbf{z}'}_{v,va}^\top {\mathbf{z}'}_{a,va} / \tau \right)}\right)
$$
 
Here, $\mathbf{z}_{v,va}$ and $\mathbf{z}_{a,va}$ are video and audio embeddings in the video-audio common space $S_{va}$, and $N$ includes all non-matching pairs in a batch.</p>

<p><strong>Multiple Instance Learning NCE (MIL-NCE) for Video-Text Pairs</strong>:  
For video-text alignment, VATT uses Multiple Instance Learning NCE (MIL-NCE). Similar to NCE, positive and negative pairs are based on temporal correspondence, but MIL-NCE accounts for noisy text data by considering multiple positive matches. The loss is defined as:


$$
\text{MIL-NCE}(\mathbf{z}_{v,vt}, \{\mathbf{z}_{t,vt}\}) = 
- \log\left(\frac{\sum_{\mathbf{z}_{t,vt} \in \mathcal{P}} \exp\left(\mathbf{z}_{v,vt}^\top \mathbf{z}_{t,vt} / \tau \right)}{\sum_{\mathbf{z}_{t,vt} \in \mathcal{P}} \exp\left(\mathbf{z}_{v,vt}^\top \mathbf{z}_{t,vt} / \tau \right) + \sum_{\mathbf{z'} \in \mathcal{N}} \exp\left(\mathbf{z'}_{v,vt}^\top \mathbf{z'}_{t,vt} / \tau \right)}\right)
$$



In this equation, $\mathbf{z}_{v,vt}$ and $\mathbf{z}_{t,vt}$ are video and text embeddings in the video-text common space $S_{vt}$, $P$ contains five text clips temporally nearest to the video clip, and $N$ includes non-matching pairs. A temperature parameter $\tau$ adjusts the softness of the objectives in distinguishing positive from negative pairs.</p>

<p><strong>Overall Loss Objective</strong>:  
The combined per-sample loss for training the entire VATT model end-to-end is:  
$$
\mathcal{L} = \text{NCE}(\mathbf{z}_{v,va}, \mathbf{z}_{a,va}) + \lambda \, \text{MIL-NCE}(\mathbf{z}_{v,vt}, \{\mathbf{z}_{t,vt}\})
$$
 
Here, $\lambda$ balances the contributions of the two losses. The model is optimized via backpropagation using the average loss over a batch of samples.</p>
</li>

</ol>
</h5> 

<h4 style="font-family: Poppins, sans-serif; font-size: 24px; font-weight: 500; margin-bottom: 20px; background-color: #6c5ce7; color: #ffffff; padding: 15px; border-radius: 10px;">
    Experiments
</h4>
<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 300;margin-bottom: 20px;">
<p> In this section, we first briefly describe the experimental setup for the pre-training and downstream
 evaluation, and then present the results and analytic interpretation of VATT in different tasks.</p>

<ol>
    <li>
       <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong> Experimental Setup</strong></h5>
    <p>    <strong>Pre-training Setup</strong>:  
VATT is pre-trained on a combination of AudioSet and a subset of HowTo100M datasets, adhering to YouTube's policies. HowTo100M provides video-audio-text triplets, while AudioSet supplies video-audio pairs. Video input consists of 32 frames sampled at 10 fps with a spatial size of $224 \times 224$, processed with random cropping, horizontal flipping, and color augmentation. Audio waveforms are sampled in sync at 48kHz, and both modalities are normalized between $[-1, 1]$. Tokenization uses patch sizes of $4 \times 16 \times 16$ for video and 128 for raw audio waveforms. Text sequences, capped at 16 tokens, are encoded as one-hot vectors with a vocabulary size of $2^{16}$. All pre-training experiments apply DropToken with a 50% drop rate. Training employs the Adam optimizer with a quarter-period cosine learning rate schedule from $1e-4$ to $5e-5$, 10k warmup steps, and 500k total steps with a batch size of 2048 (512 for exploration experiments). Common space projections use dimensions $d_{va} = 512$ for video-audio space $S_{va}$ and $d_{vt} = 256$ for video-text space $S_{vt}$, with a temperature $\tau = 0.07$ and loss weight $\lambda = 1$ in the combined loss equation. VATT experiments with four network sizes: Medium (155M parameters) for the modality-agnostic variant (VATT-MA), and three modality-specific variants: Base-Base-Small (BBS; 197M), Medium-Base-Small (MBS; 264M), and Large-Base-Small (LBS; 415M). Pre-training an MBS VATT with batch size 2048 on 256 TPUs (v3) takes less than 3 days, and with batch size 512, it takes under 1 day.
</p>
<p>
<strong>Downstream Evaluation</strong>:  <br>
Pre-trained VATT models are evaluated on four major downstream tasks across 10 datasets:  
<ul>
<li> <strong>Video Action Recognition</strong>: UCF101, HMDB51, Kinetics-400, Kinetics-600, and Moments in Time.</li>  
<li><strong>Audio Event Classification</strong>: ESC50 and AudioSet.</li>  
<li><strong>Zero-Shot Text-to-Video Retrieval</strong>: YouCook2 and MSR-VTT, assessing video-text common space quality.  </li>
<li><strong>Image Classification</strong>: ImageNet, testing vision backbone transferability. </li>
</ul><br>
For smaller datasets like HMDB51, UCF101, and ESC50, only a linear classifier is trained on top of frozen pre-trained backbones. Exploration experiments report linear classification accuracy and zero-shot retrieval metrics. Detailed dataset descriptions and experimental setups are provided in the paper's appendix.
</p>
    </li>

<li>
       <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong>Results </strong></h5>

<ol>
    <li>
        <h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong> Fine-tuning for video action recognition </strong></h5>
        <p>VATT's vision Transformer was fine-tuned on Kinetics-400, Kinetics-600, Moments in Time, and Kinetics-700, using final checkpoints from four pre-training settings: three modality-specific variants (LBS, MBS, BBS) and one modality-agnostic (Medium). The model achieves top-1 accuracies of 82.1% on Kinetics-400, 83.6% on Kinetics-600, 41.1% on Moments in Time, and 72.7% on Kinetics-700, surpassing previous state-of-the-art models like TimeSFormer, which rely on supervised pre-training. Notably, VATT is the first vision Transformer pre-trained from scratch using self-supervision on multimodal videos to achieve these records. A variant trained from scratch without pre-training yields only 26.4% top-1 and 51.8% top-5 accuracy on Kinetics-400, highlighting the efficacy of VATT's pre-training strategy. Additionally, the modality-agnostic VATT-MA-Medium performs on par with the modality-specific VATT-Base, suggesting the potential for a unified Transformer backbone across modalities.</p>
    </li>

By replicating input images four times to mimic single-frame video clips, the model achieves a top-1 accuracy of 78.7%, comparable to ViT's 79.9% (which uses supervised pre-training on large-scale image data) and significantly better than 64.7% from training the same Transformer from scratch. Performance is affected by noisy text transcripts, suggesting future exploration of higher-quality text sources.</p>

<li>
<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong> Feature Visualization </strong></h5>
    <p>Feature representations of fine-tuned VATT models on Kinetics-400 were visualized using t-SNE, showing better separation compared to a model trained from scratch. Modality-agnostic and modality-specific features show no clear difference. Pre-trained representations from YouCook2 clips at tokenization and common space projection stages indicate that modality-agnostic backbones perceive modalities as symbols of the same concept, akin to unified language models in NLP. Kernel Density Estimation (KDE) of pairwise similarities confirms VATT's ability to separate positive and negative video-text pairs in its output space, verifying the semantic common space learning.</p>

<h4 style="font-family: Poppins, sans-serif; font-size: 24px; font-weight: 500; margin-bottom: 20px; background-color: #6c5ce7; color: #ffffff; padding: 15px; border-radius: 10px;">
    Reflections on VATT: Surprises and Scope for Improvement
</h4>

<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 300;margin-bottom: 20px;">
    <ol>
        <li>
<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong> What Surprised Me? </strong></h5>
    <p>Studying the Video-Audio-Text Transformer (VATT) revealed several unexpected outcomes that highlight the power and potential of Transformer architectures in multimodal learning. One of the most surprising aspects was VATT's exceptional transferability across domains, particularly in image classification. Despite being pre-trained on multimodal video data, fine-tuning VATT's vision Transformer on ImageNet achieved a top-1 accuracy of 78.7%, remarkably close to Vision Transformer's (ViT) 79.9%, which relies on supervised pre-training with large-scale image data. This near-parity, despite the domain gap between dynamic video and static image data, underscores the robustness and generalizability of self-supervised multimodal pre-training in learning visual priors that transcend specific data types. Additionally, I was astonished by the performance of the modality-agnostic backbone (VATT-MA-Medium), which performed on par with modality-specific models in tasks like video action recognition and audio event classification. The idea that a single Transformer with shared weights can effectively handle video, audio, and text modalities-perceiving them as different symbols of the same underlying concept-challenges conventional assumptions about the need for specialized architectures and hints at a future of unified AI models.</p>
<p>Another surprising element was VATT's ability to outperform CNN-based models in audio event classification (achieving a mean Average Precision of 39.4% on AudioSet) while operating directly on raw waveforms without handcrafted features. This result defies the long-standing dominance of CNNs in audio processing and demonstrates that Transformers, with self-attention mechanisms, can capture temporal dependencies in raw data more effectively than traditional approaches. Equally surprising is how VATT performs as well as, or better than, CNN-based models in video action recognition tasks, achieving top-1 accuracies of 82.1% on Kinetics-400 and 83.6% on Kinetics-600, surpassing established CNN architectures. This challenges the entrenched belief that CNNs, with their inductive biases like locality and translation invariance, are inherently superior for vision tasks, proving that convolution-free Transformers can excel when trained on large-scale unlabeled multimodal data. These unexpected successes highlight how far self-supervised learning and Transformer architectures have come in redefining multimodal AI.</p>
</li>

<li>
<h5 style="font-family: Poppins, sans-serif; font-size: 18px; font-weight: 500;margin-bottom: 20px;"><strong>Scope for Improvement </strong></h5>
    <p>While VATT achieves groundbreaking results, several areas offer opportunities for enhancement. Improving text data quality by using higher-quality sources like curated captions could boost zero-shot retrieval performance on datasets like YouCook2, addressing issues with noisy transcripts. Computational efficiency remains a challenge despite DropToken, as training large models is resource-intensive; exploring efficient attention mechanisms like Performer could make VATT more accessible. Extending the framework to include modalities like depth or tactile data could test the modality-agnostic concept further. Additionally, integrating a Mixture-of-Experts approach, as suggested by model activation patterns, might enhance multimodal capacity. Finally, improving robustness to data imbalance and noise in multimodal alignments could stabilize learning, ensuring better representation quality across diverse datasets.</p>
</li>
    </ol>
